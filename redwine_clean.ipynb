```r
library(tidyverse)
library(repr)
library(scales)
source("tests.R")
source("cleanup.R")
```

```r
wine <- read_csv('winequality-red.csv')
wine
```

```r
mean1<- group_by(wine, quality) |> 
    summarize(mean_fixed_acidity = mean(fixed_acidity))
mean1
mean_fixed_acidity_plot <- ggplot(mean1, aes(x=quality, y=mean_fixed_acidity)) + 
    geom_bar(stat="identity") + 
    labs(x='quality', y='mean of fixed acidity')

mean_fixed_acidity_plot
```

```r
mean8<- group_by(wine, quality) |> 
    summarize(mean_density = mean(density))
mean8 <- mutate(mean8, mean_density_scaled = mean_density - 0.99)
mean8
mean_density_plot <- ggplot(mean8, aes(x=quality, y=mean_density)) + 
    geom_bar(stat="identity") + 
    labs(x='quality', y='mean of density)') +
    geom_hline(yintercept = 0.996, linetype = "dashed", color = "red", linewidth = 0.5)
mean_density_plot

mean_density_scaled_plot <- ggplot(mean8, aes(x=quality, y=mean_density_scaled)) + 
    geom_bar(stat="identity") + 
    labs(x='quality', y='scaled mean density') 
mean_density_scaled_plot
```
