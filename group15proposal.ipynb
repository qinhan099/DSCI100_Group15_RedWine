**Introduction**

The data set proposed describes the red wine variants of the Portuguese "Vinho Verde" wine which was collected by researchers (Cortez et al., 2009). While every data point is a red wine, they different in certain aspects and quality. This data is a collection of eleven input variables pertaining to different components of wine as scientifically calculated values that were determined based on a series of physicochemical tests. The output variable is a quality score on a scale of 0 to 10 that consists of integer values categorizing each wine based on their inputs. Note, the authors of the data arbitarily created score values for each case of red wine depending on the researched chemical scores. The data consist of input variables fixed acidity, volatile acidity, citric acid, residual sugar, chlorides, free sulfur dioxide, total sulfur dioxide, density, pH, sulfates, and alcohol. The quality scores are given arbitarily by the author based on the chemical scores. 

The question that is going to be answered through this research are:
- Is there any relationship between each chemical score and the quality of the wine? If so, what would it look like?
-  Given its chemical scores, what would the quality score of a specific wine be?

```r
# Use as our discussion for the possibility to reach a multiple linear regression model -> Because our classification model does not provide a solid and 
# concrete accuracy, we want to try out whether linear regression model is going to work or not.
# Multi-factor linear regression to predict the quality of the wine.
# Using the testing data set and the training data set we've already dissected previously
# wine_split <- initial_split(wine, prop = 0.75, strata = quality)
# wine_train <- training(wine_split)
# wine_test <- testing(wine_split) 
# Filter our training data using the factors we found that has the obvious relationships with Quality.
# Quality Correlation Factors: Volatile_acidity, Critic_acid, chlorides, free_sulfur_dioxide, density, Sulphates

wine_lm_spec <- linear_reg() |> set_engine("lm") |> set_mode("regression")
wine_linear_recipe <- recipe(quality ~ volatile_acidity + citric_acid 
                             + chlorides + free_sulfur_dioxide + density + sulphates, data= wine_train)

print(wine_lm_spec)
print(wine_linear_recipe)
```

Use as our discussion for the possibility to reach a multiple linear regression model -> Because our classification model does not provide a solid and concrete accuracy, we want to try out whether linear regression model is going to work or not. Filter our training data using the factors we found that has the obvious relationships with Quality.

```r
wine_linear_fit <- workflow() |> add_recipe(wine_linear_recipe) |> add_model(wine_lm_spec) |> fit(data= wine_train)
wine_linear_fit

lm_prediction_testing <- wine_linear_fit |> predict(wine_test) |> bind_cols(wine_test)
lm_prediction_testing
```

```r
# 1. After we've made prediction using the factors that we've explored, let's find the lm-rspe difference from the testing data set.
wine_lm_rmspe <- lm_prediction_testing |> metrics(truth=quality, estimate=.pred)|>
            filter(.metric=='rmse') |> pull(.estimate)
wine_lm_rmspe

# 2. Let's visualize the difference by plotting them in the same data visualization line to see the actual difference.

wine_difference_plot <- ggplot()
```

```r
# 3. Maybe let's use rounding to make the prediction makes more sense and thus decreasing the error range for rmspe.
wine_rounded_predicted <- lm_prediction_testing |> mutate(.pred = round(.pred))
wine_rounded_predicted
# 4. Now let's calculate for the LM_RMSPE again.
rounded_lm_rmspe <- wine_rounded_predicted |> metrics(truth=quality, estimate=.pred) |> filter(.metric=='rmse') |> pull(.estimate)
rounded_lm_rmspe
# After We've used round to account for the RMSE(Root Mean Squared Error), we've found that it produced larger difference.
# With this condition, we can conclude that our estimation factors might included some factors that caused a mild prediction error
# In our experiment.
```
